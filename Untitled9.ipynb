```python

# Install gradio if it's not already installed
!pip install gradio

import gradio as gr
from transformers import pipeline

# Emotion-to-emoji mapping
emotion_emoji_map = {
    "joy": "😄", "sadness": "😢", "anger": "😠", "fear": "😱", "disgust": "🤢",
    "surprise": "😲", "love": "❤️", "admiration": "👏", "amusement": "😂",
    "gratitude": "🙏", "nervousness": "😬", "optimism": "🌟", "pride": "🏅",
    "disappointment": "😞", "relief": "😌", "caring": "🤗", "remorse": "😔",
    "confusion": "😕", "realization": "💡", "neutral": "😐"
}

# Load emotion classifier pipeline
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base")

# Function to analyze emotion
def analyze_emotion(text):
    result = classifier(text)[0]
    emotion = result['label']
    score = result['score']
    emoji = emotion_emoji_map.get(emotion.lower(), "❓")
    return f"Detected Emotion: {emotion} {emoji} (Confidence: {score:.2f})"

# Gradio interface
gr.Interface(
    fn=analyze_emotion,
    inputs=gr.Textbox(lines=2, placeholder="Enter a social media message..."),
    outputs="text",
    title="Emotion Detection with Emoji",
    description="Enter a message to detect the emotion and view its corresponding emoji."
).launch()
```
